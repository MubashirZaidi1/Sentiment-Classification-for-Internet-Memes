```python
X = df.text_corrected.values.astype(str)
y = df.sents
le = LabelEncoder()
y = le.fit_transform(y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4) # 70% training and 30% test

vectorizer = CountVectorizer()
X_trainn = vectorizer.fit_transform(X_train)
X_testt = vectorizer.transform(X_test)

clf = DecisionTreeClassifier( min_samples_split=5,
    min_samples_leaf=5,
    min_weight_fraction_leaf=0.2,)
clf.fit(X_trainn,y_train)
y_pred=clf.predict(X_testt)
print('Accuracy:', accuracy_score(y_test, y_pred))
print("F1 Score: ",f1_score(y_test, y_pred,average='macro')) 
print(classification_report(y_test, y_pred))
```
